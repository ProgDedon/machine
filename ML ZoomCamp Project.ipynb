### Encoding

```python
dfencode = pd.get_dummies(df, prefix = ["schoolsup", "famsup", "paid", "activities", "nursery", "higher", "internet", "romantic", "school", "address", "Pstatus", "famsize", "Mjob", "Fjob", "reason", "guardian"], columns=["schoolsup", "famsup", "paid", "activities", "nursery", "higher", "internet", "romantic", "school", "address", "Pstatus", "famsize", "Mjob", "Fjob", "reason", "guardian"])
```

```python
pd.set_option('display.max_columns', None)
dfencode.head()
```

```python
from sklearn.preprocessing import LabelEncoder
multicols = ["Medu", "Fedu"]
dfencode[multicols] = dfencode[multicols].apply(LabelEncoder().fit_transform)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_map = [["very bad", "bad", "average", "good", "very good"]]
```

```python
encoder = OrdinalEncoder(categories = ord_map)
dfencoded= encoder.fit_transform(df[["health"]])
dfencode["health_encode"] = dfencoded
```

```python
dfencode.health.unique().tolist()
```

```python
dfencode = dfencode.drop("sex", axis=1)
dfencode = dfencode.drop("health", axis=1)
```

### Model Building

##### Linear Regression

```python
sns.regplot(x='G1', y='Grade', data=dfencode)
```

```python
sns.regplot(x='G2', y='Grade', data=dfencode)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = dfencode["Grade"]
x = dfencode.drop("Grade", axis=1)
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(x, y)
```

```python
print(linreg.intercept_, linreg.coef_)
```

```python
from sklearn.model_selection import train_test_split
model= LinearRegression()
```

```python
Y = dfencode["Grade"]
X = dfencode.drop("Grade", axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model.fit(X_train, Y_train)
```

```python
features = X_train.columns  
coefficients = model.coef_
for feature, coef in zip(features, coefficients):
    print(f"{feature}:\t\t\t {coef}")
```

```python
selected_features = [feature for feature, coef in zip(features, coefficients) if coef > 0.5 or coef < -0.5]
```

```python
selected_features
```

```python
X_train_sel = X_train[selected_features]
```

```python
linregmodel = LinearRegression()
```

```python
linregmodel.fit(X_train_sel, Y_train)
```

```python
Y_sel_pred = linregmodel.predict(X_train_sel)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_train, Y_sel_pred)
```

```python
Y_test_sel_pred = linregmodel.predict(X_test[selected_features])
r2_score(Y_test, Y_test_sel_pred)
```

##### Random Forest

##### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
# recall
Y = dfencode["Grade"]
X = dfencode.drop("Grade", axis=1)
```

```python
# function to make prediction
Y_pred_en = clf_entropy.predict(X_test)
print(Y_pred_en)
```
